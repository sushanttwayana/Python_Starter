```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

```

```python
k = np.arange(0,17)
print(k)
```

```python
pmf = poisson.pmf(k, mu=7)
pmf = np.round(pmf, 5)
print(pmf)
```

```python
plt.plot(k, pmf, color = 'r',marker ='x')
# k is x-axis and pmgf is y axis
plt.xlabel('k')
plt.ylabel('Probability')
plt.show()

# 'b' - blue
# 'g' - green
# 'r' - red
# 'c' - cyan
# 'm' - magenta
# 'y' - yellow
# 'w' - white
```

```python
cdf = poisson.cdf(k , mu =7)
cdf = np.round(cdf,3)
print(cdf)
```

```python
# Create a figure with a specific size
plt.figure(figsize=(8, 6))

plt.title('Cumulative Probability Distribution')
plt.grid(True)

plt.plot(k, cdf, color = 'm', marker= 'o', label = 'CDF')
plt.legend()

plt.xlabel('k')
plt.ylabel('Cumulative Probability')
plt.show()
```

```python
#Parameters for the exponential distribution
lamda_param = 0.2 # Rate parameter (Inverse of mean time between events)

#Generate exponential waiting times
num_waiting_times = 1000
waiting_times = np.random.exponential(scale = 1/lamda_param, size = num_waiting_times)

#plot the histogram of waiting times
plt.hist(waiting_times, bins = 30, density = True, alpha = 0.6, color = 'blue',edgecolor = 'black', label ='Generated Data')

#plot the PDF of the exponential function
x = np.linspace(0, max(waiting_times), 100)
pdf = lamda_param * np.exp (-lamda_param * x)
plt.plot(x, pdf, 'r', label = 'Exponential PDF')

plt.title('Exponential Distribution of waiting times')
plt.xlabel('Waiting Time')
plt.ylabel('Probabilty Density')
plt.legend()
plt.plot()
```
