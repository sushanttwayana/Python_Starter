```python
#Estimation of pi

import random
import matplotlib.pyplot as plt
import numpy as np

def estimate_pi(num_points):
  points_inside_circle = 0
  points_inside_x = []
  points_inside_y = []
  points_outside_x = []
  points_outside_y = []

  for _ in range(num_points):
    x = random.uniform(0,1)
    y = random.uniform(0,1)

    distance = x**2 + y**2
    if distance <= 1:
      points_inside_circle += 1
      points_inside_x.append(x)
      points_inside_y.append(y)
    else:
      points_outside_x.append(x)
      points_outside_y.append(y)

  pi_estimate = 4*points_inside_circle / num_points
  return pi_estimate, points_inside_x, points_inside_y, points_outside_x, points_outside_y


num_points = 1550 #Number of random points to generate
estimated_pi, inside_x, inside_y, outside_x, outside_y = estimate_pi(num_points)

print(f"Estimated value of n: {estimated_pi}")


#plotting the scatter of points and the quarter circle

plt.figure(figsize=(8,8))
plt.scatter(inside_x, inside_y, color='blue', label='Inside Quarter Circle')
plt.scatter(outside_x, outside_y, color='red', label='Outside Quarter Circle')
plt.plot(np.linspace(0, 1, 100), np.sqrt(1 - np.linspace(0, 1, 100)**2), color='green', label='Quarter Circle')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Monte Carlo Simulation: Estimating n')
plt.legend()
plt.axis('equal')  #Equal scaling for x and y axes
plt.show()

```
